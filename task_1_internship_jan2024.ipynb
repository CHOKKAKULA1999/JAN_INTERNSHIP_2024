```python
Indf_copy['DOB'] = pd.to_datetime(Indf_copy['DOB'])
```

```python
Indf_copy.dtypes
```

```python
Indf_copy['YEAR'] = Indf_copy['DOB'].dt.year
Indf_copy['MONTH'] = Indf_copy['DOB'].dt.month
Indf_copy['DAY'] = Indf_copy['DOB'].dt.day
```

```python
Indf_copy.drop([ 'MONTH', 'DAY'], axis=1, inplace=True)
```

```python
Indf_copy['MonthOfBirth'] = Indf_copy['DOB'].dt.strftime('%B')  # %B gives the full month name
Indf_copy['DayOfBirth'] = Indf_copy['DOB'].dt.strftime('%A')   # %A gives the full day name
```
